```python
# Transpose data and merge class label data condition
# df_t = exp_data_UC_H.set_index('Gene.Symbol').T.reset_index()

# df_t.rename(columns={'index': 'SampleID'}, inplace=True)
# merged_df = pd.merge(df_t,condition, on="SampleID")


# df = exp_data_doses.set_index('Gene.Symbol').T.reset_index()
# df.rename(columns={'index': 'SampleID'}, inplace=True)
# m_df = pd.merge(df,conditions, on="SampleID")


# df = expression_data_rem.set_index('Gene.Symbol').T.reset_index()

# df.rename(columns={'index': 'SampleID'}, inplace=True)
# merged_df_rem = pd.merge(df,metadata_rem, on="SampleID")

# # Select features into X from the merged dataset, and y is the condition of patients 
# y=merged_df_rem["Remission"]
# X=merged_df_rem.drop(columns=["SampleID","Condition"])



def preprocess_expression_data(expression_data, metadata, id_col="SampleID", target_col="Condition", index="Gene.Symbol", drop_cols=None):
    """
    Transposes the expression dataset, merges it with metadata, and extracts features (X) and target labels (y).

    Parameters:
    - expression_data (pd.DataFrame): Expression data with 'Gene.Symbol' as index column.
    - metadata (pd.DataFrame): Metadata containing subject labels.
    - id_col (str): Column name for subject/sample IDs.
    - target_col (str): Column name for condition labels.

    Returns:
    - X (pd.DataFrame): Feature matrix with gene expression values.
    - y (pd.Series): Target labels for patient conditions.
    - merged_df (pd.DataFrame): Merged expression and metadata dataframe.
    """
    if drop_cols is None:
        drop_cols = []
    # Transpose data and reset index
    df_t = expression_data.set_index(index).T.reset_index()

    # Rename the index column to match metadata
    df_t.rename(columns={'index': id_col}, inplace=True)

    # Merge transposed data with metadata
    merged_df = pd.merge(df_t, metadata, on=id_col)

    # Extract target labels
    y = merged_df[target_col]

    # Extract features (excluding SampleID and target column)
    X = merged_df.drop(columns=[id_col, target_col]+drop_cols)

    return X, y, merged_df

# drop = ["Condition","Week"]
# X,y,merged_df = preprocess_expression_data(expression_data_rem,metadata_rem,id_col="SampleID", target_col="Remission",index="Gene.Symbol",drop_cols = drop)
# print(X.head())
# X,y,merged_df = preprocess_expression_data(exp_data_doses,conditions,id_col="SampleID",target_col = "Condition",index="Gene.Symbol",drop_cols = None)
# print(X.head())
# print(y.head())

# X,y,merged_df = preprocess_expression_data(exp_data_UC_H,condition,id_col="SampleID",target_col = "Condition",index="Gene.Symbol",drop_cols = None)





# def filter_for_sig_genes(expression_df, degs_file,num_genes=None,id_col,target_col):
#     if num_genes is None:
#         degs = degs_file.iloc[:, 0].tolist() 
#     else:
#         degs = degs_file.iloc[:num_genes, 0].tolist()

#     degs = degs_file.iloc[:num_genes,0].tolist()
#     filt_expression_data = expression_df[[id_col,target_col]]
#     deg = expression_df[degs]
#     filt_data = pd.concat([filt_expression_data,deg],axis=1)
#     return filt_data
    
# filt_data = filter_for_sig_genes(merged_df,genes,50)     



# print(m_df["Condition"].value_counts())
# uc_subjects = merged_df[merged_df["Condition"] == "UC"]
# healthy_subjects = merged_df[merged_df["Condition"] == "healthy"]

# # Randomly selecting 18 subjects from each group
# uc_sample = uc_subjects.sample(n=18, random_state=42)
# healthy_sample = healthy_subjects.sample(n=18, random_state=42)

# final_sample = pd.concat([uc_sample, healthy_sample])

# final_sample = final_sample.sample(frac=1, random_state=42).reset_index(drop=True)
    
# Select features and class label after DEGs analysis filtering
# y = final_sample["Condition"]
# X = final_sample.drop(columns=["SampleID","Condition"])




# # # Filter the expression data to contain rows with DEGs only
# # df = exp_data_UC_H[exp_data_UC_H["Gene.Symbol"].isin(genes["GeneID"].values)]
# # df_transposed = df.melt(id_vars="Gene.Symbol",var_name="SampleID",value_name="Gene expression")
# # merged_df = pd.merge(df_transposed,condition, on="SampleID")
# # # print(merged_df.head())
# #
# # all_genes_transposed = exp_data_UC_H.melt(id_vars="Gene.Symbol",var_name="SampleID",value_name="Gene expression")
# # merged_all_genes = pd.merge(all_genes_transposed,condition, on="SampleID")

# df_t = exp_data_UC_H.set_index('Gene.Symbol').T.reset_index()
# df_t.rename(columns={'index': 'SampleID'}, inplace=True)
# merged_df = pd.merge(df_t,condition, on="SampleID")
```

```python
# Experiments with one feature (gene) training, outputs and plots
X,y,merged_df = preprocess_expression_data(exp_data_UC_H,condition,id_col="SampleID", target_col="Condition",index="Gene.Symbol",drop_cols = None)

# Normalise data using Z-score normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

genes = X.columns[:10]
results = []

random_genes = np.random.choice(X.columns, size=10, replace=False)
for gene in random_genes:
    X_gene = X_scaled[:, X.columns.get_loc(gene)].reshape(-1, 1)  # Get one gene at a time
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_gene, y, test_size=0.25, stratify = y, random_state=42)
    
    print("Train class distribution:", {label: list(y_train).count(label) for label in set(y)})
    print("Test class distribution:", {label: list(y_test).count(label) for label in set(y)})


    knn = KNeighborsClassifier(n_neighbors = 3)

    # Train the classifier
    knn.fit(X_train, y_train)

    # Make predictions
    y_test_pred = knn.predict(X_test)

    # Evaluate the classifier
    accuracy = round(accuracy_score(y_test, y_test_pred),5)
    b_acc_test = round(metrics.balanced_accuracy_score(y_test, y_test_pred),5)
    precision_score_test = round(metrics.precision_score(y_test,y_test_pred,average = "weighted"),5)
    recall_score_test = round(metrics.recall_score(y_test,y_test_pred,average = "weighted"),5)
    f1_score_test = round(metrics.f1_score(y_test,y_test_pred,average = "weighted"),5)
     # Store results
    results.append({
        "Gene": gene,
        "Accuracy": accuracy,
        "Balanced Accuracy": b_acc_test,
        "Precision": precision_score_test,
        "Recall": recall_score_test,
        "F1 score:" :  f1_score_test
            
    })

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Save to CSV
results_df.to_csv("random_gene_KNN_classification_results.csv", index=False, float_format="%.5f")

# Display table
print(results_df)

    
print(f"""For {gene}: accuracy {accuracy:.4f},balanced_accuracy: {b_acc_test:.4f},precison score: {precision_score_test:.4f},
recall score: {recall_score_test:.4f}""")

# Initialize a figure for plotting
plt.figure(figsize=(15, 10))
distance_tables = []

# Plot distances for the first 5 test cases (you can change this number to plot more test cases)
for i, test_case in enumerate(X_test[:5]):  # Change the number here to plot more test cases
    test_case = test_case.reshape(1, -1)  # Reshape to 2D array (one test case)
    
    # Get the distances and indices of the nearest neighbors
    distances, indices = knn.kneighbors(test_case)
    
    # Get the classes of the test case and its nearest neighbors
    test_case_class = y_test.iloc[i]  # The class of the test case
    neighbors_classes = y_train.iloc[indices[0]] # The classes of the nearest neighbors

    # Create a DataFrame to display the distances and neighbor classes
    distance_table = pd.DataFrame({
        'Neighbor #': range(1, len(distances[0]) + 1),
        'Distance': distances[0],
        'Neighbor Class': neighbors_classes
    })
    distance_tables.append(distance_table)


  # Get the classes of the test case and its nearest neighbors

    # Check if any distance is zero and print a warning if found
    if any(distances[0] == 0):
        print(f"Warning: Zero distance found for Test Case {i+1}")
    
    
    # Get the predicted class for the test case
    predicted_class = knn.predict(test_case)[0]
    
    # Check if the classification is correct
    is_correct = predicted_class == test_case_class
    # Create a subplot for each test case
    ax = plt.subplot(2, 3, i+1)  # 2 rows, 3 columns, plot the ith test case
    ax.plot(range(1, len(distances[0]) + 1), distances[0], linestyle='-', color='black', label=f'Test Case {i+1} - {test_case_class}')

    for j, (dist, neighbor_class) in enumerate(zip(distances[0], neighbors_classes)):
        color = 'blue' if neighbor_class == 'Healthy' else 'orange'
        ax.plot(j+1, dist, 'o' ,color=color)

    # Add annotation for correct/incorrect classification
    classification_status = "Correct" if is_correct else "Incorrect"
    ax.text(len(distances[0]) + 0.2, np.mean(distances[0]), f'Predicted: {predicted_class}\n{classification_status}',
             color='black', ha='left', va='center')

    # Labels and title
    ax.set_title(f'Test Case {i + 1}')
    ax.set_xlabel('Neighbor #')
    ax.set_ylabel('Distance')
    ax.set_xticks(range(1, len(distances[0]) + 1))


# Create a legend
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=8, label='Healthy'),
           plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='orange', markersize=8, label='UC')]
ax.legend(handles=handles, loc='upper right', bbox_to_anchor=(2, 1))

# Adjust layout and save
plt.tight_layout()
plt.savefig('nearest_neighbors_distances.png', dpi=300)
plt.show()


# Print the distance tables for each test case
for i, table in enumerate(distance_tables):
    print(f"Distance table for Test Case {i+1}:")
    print(table)
    print("\n")
```

```python
# Define classifiers
knn = KNeighborsClassifier()
knn1 = KNeighborsClassifier(n_neighbors = 10)
knn2 = KNeighborsClassifier(n_neighbors = 3)

dt = DecisionTreeClassifier()
dt1 = DecisionTreeClassifier(max_depth = 100)
dt2 = DecisionTreeClassifier(max_depth = 25)
dt3 = DecisionTreeClassifier(max_depth = 4)

svc_lin = SVC(kernel = 'linear')
svc_quad = SVC(kernel = 'poly',degree=2)
svc_cubic = SVC(kernel = 'poly',degree=3)

rfc = RotationForestClassifier(n_estimators=100)

classifiers = [knn,knn1,knn2,dt,dt1,dt2,dt3,svc_lin,svc_quad,svc_cubic,rfc]



```

```python
# Normalise 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 1: Encode the target variable (y)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Perform stratified split (75% train, 25% test) and ensure the same class proportions in the test and train sets 
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.25, stratify=y, random_state=42)

print("Train class distribution:", {label: list(y_train).count(label) for label in set(y_encoded)})
print("Test class distribution:", {label: list(y_test).count(label) for label in set(y_encoded)})



output_dir = '../../results/tables'

def get_model_params(model, params):
    model_params = {}
    for param in params:
        if hasattr(model, param):  # Check if the model has the parameter
            model_params[param] = getattr(model, param)  # Get the parameter value
    return model_params

def evaluate_classifiers(classifiers_lst, X_train, y_train, X_test, y_test, file_name = "default",params=None):
    results = []
    for idx,model in enumerate(classifiers_lst):
        start_time = time.time()

        # Print progress feedback
        print(f"Evaluating {model.__class__.__name__} ({idx + 1}/{len(classifiers)})...")

        # Check if it's a multi-class or binary classification problem

        model.fit(X_train, y_train)
        
        # Training set predictions and metrics
        y_train_pred = model.predict(X_train)
        acc_train = round(metrics.accuracy_score(y_train, y_train_pred),5)
        b_acc_train = round(metrics.balanced_accuracy_score(y_train, y_train_pred),5)
        precision_score_train = round(metrics.precision_score(y_train,y_train_pred,average = "weighted"),5)
        recall_score_train = round(metrics.recall_score(y_train,y_train_pred,average = "weighted"),5)
        f1_score_train = round(metrics.f1_score(y_train,y_train_pred,average = "weighted"),5)
        print(f"""{model.__class__.__name__} accuracy on training set: {acc_train:.5f}, balanced accuracy: {b_acc_train:.5f},
        precision score: {precision_score_train:.5f}, f1 score: {f1_score_train:.5f}""")
        
        # Test set predictions and metrics
        y_test_pred = model.predict(X_test)
        acc_test = round(metrics.accuracy_score(y_test, y_test_pred),5)
        b_acc_test = round(metrics.balanced_accuracy_score(y_test, y_test_pred),5)
        precision_score_test = round(metrics.precision_score(y_test,y_test_pred,average = "weighted"),5)
        recall_score_test = round(metrics.recall_score(y_test,y_test_pred,average = "weighted"),5)
        f1_score_test = round(metrics.recall_score(y_test,y_test_pred,average = "weighted"),5)

        
        print(f"""{model.__class__.__name__} accuracy on test set: {acc_test:.5f}, balanced accuracy: {b_acc_test:.5f}, 
        precision score: {precision_score_test:.5f}, recall score: {recall_score_test:.5f}, f1 score: {f1_score_test:.5f}""")

        # Extract classifier parameters dynamically
        model_params = get_model_params(model,params)
        
        
        # Store results with classifier name, metrics, and selected parameters to display
        row = {
            "Classifier": model.__class__.__name__,
            "Train Accuracy": acc_train,
            "Test Accuracy": acc_test,
            "Train Balanced Accuracy": b_acc_train,
            "Test Balanced Accuracy": b_acc_test,
            "Train Precision": precision_score_train,
            "Test Precision": precision_score_test,
            "Train Recall": recall_score_train,
            "Test Recall": recall_score_test,
            "Train F1 Score": f1_score_train,
            "Test F1 Score": f1_score_test
        }

        # Merge selected hyperparameters into the row
        row.update(model_params)


        # Append row to results list
        results.append(row)

        # End time tracking for this model
        end_time = time.time()
        time_taken = end_time - start_time
        
        # Convert time to minutes and seconds
        minutes = int(time_taken // 60)
        seconds = int(time_taken % 60)
        
        # Print time taken for the model
        print(f"Time taken for {model.__class__.__name__}: {minutes} minutes {seconds} seconds\n")

        # Convert results list to DataFrame
        results_df = pd.DataFrame(results)

        # Define output file name dynamically
        filename = f"classifier_evaluation_results_{file_name}.csv"
        file_path = os.path.join(output_dir, filename)
    
        # Save results to CSV
        results_df.to_csv(file_path, index=False)
    
        print(f"Results saved to {file_path}")

    return results_df  
params = ["kernel", "max_depth","n_neighbors","degree","n_estimators"]

# evaluate_classifiers(classifiers, X_train, y_train, X_test, y_test, file_name="UC_130_H_no_placebo_all_genes", params=params)
evaluate_classifiers(classifiers, X_train, y_train, X_test, y_test, file_name="remission_UC_placebo_W8_all_genes", params=params)
# evaluate_classifiers(classifiers, X_train, y_train, X_test, y_test, file_name="all_doses_UC_H_all_genes", params=params)


# for i in classifiers:
#     evaluate_model(i,X_train,y_train,X_test,y_test)
```
